```python
# Obtain information about all liquidity pools on a specified blockchain.
import requests

headers = {
    'accept': 'application/json',
    'x-api-key': '',
}

params = {
    'sort': 'creationTime',
    'order': 'desc',
    'from': '2024-03-12T23:00:00.000Z',
    'to': '2024-03-15T23:00:00.000Z',
    'pageSize': '50',
}

response = requests.get('https://public-api.dextools.io/trial/v2/pool/ether', params=params, headers=headers)
response.json()
```
